#Product Matching

```python
import numpy as np
import pandas as pd
```

```python
path_1="/content/amz_com-ecommerce_sample.csv"
path_2 ="/content/flipkart_com-ecommerce_sample.csv"
amazon_data=pd.read_csv(path_1,encoding="latin-1")
flipkart_data=pd.read_csv(path_2,encoding="latin-1")
```

```python
amazon_data
```

```python
df=pd.DataFrame()
df['new'] = np.where((amazon_data['product_name'] == flipkart_data['product_name']),amazon_data["product_name"],flipkart_data["product_name"])
```

```python
df
```

```python
Match=df.drop_duplicates()
```

```python
Match
```

```python
final=pd.DataFrame()
final["Product Name in Flipkart"]= Match["new"]
```

#Adding all Flipkart Prices in Final DF

```python
for product in final["Product Name in Flipkart"]:
     if (product == flipkart_data["product_name"]).any():
        final["Retail Price in Flipkart"]= flipkart_data["retail_price"]
        final["Discounted Price in Flipkart"] = flipkart_data["discounted_price"]
```

```python
final
```

#Adding all Amazon Prices in Final DF

```python
final["Product Name in Amazon"]= Match["new"]
for product in final["Product Name in Amazon"]:
     if (product == amazon_data["product_name"]).any():
        final["Retail Price in Amazon"]= amazon_data["retail_price"]
        final["Discounted Price in Amazon"] = amazon_data["discounted_price"]
```

```python
final
```
